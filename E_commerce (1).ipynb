```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style='whitegrid')
```

### Order by Delivery Status

```python
plt.figure(figsize=(7, 4))
sns.countplot(data=stud, x='delivery_status', palette='Set2')
plt.title('Count of Orders by Delivery Status')
plt.xlabel('Delivery Status')
plt.ylabel('Number of Orders')
plt.xticks(rotation=45)
plt.show()
```

### Top cities by number of Deliveries

```python
top_cities = stud[stud['delivery_status'] == 'Delivered']['customer_city'].value_counts().head(10)
plt.figure(figsize=(8, 4))
sns.barplot(x=top_cities.values, y=top_cities.index, palette='Blues_d')
plt.title('Top 10 Cities by Number of Deliveries')
plt.xlabel('Number of Deliveries')
plt.ylabel('City')
plt.show()
```

### Distribution of Total Price

```python
plt.figure(figsize=(10, 4))
sns.histplot(stud['total_price'], bins=30, kde=True, color='purple')
plt.title('Distribution of Total Price')
plt.xlabel('Total Price')
plt.ylabel('Frequency')
plt.show()
```

### Total price by Payment type

```python
plt.figure(figsize=(10, 4))
sns.boxplot(data=stud, x='payment_method', y='total_price', palette='Set3')
plt.title('Total Price by Payment Type')
plt.xlabel('Payment Type')
plt.ylabel('Total Price')
plt.xticks(rotation=45)
plt.show()
```
